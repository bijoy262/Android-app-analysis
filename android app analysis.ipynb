```python
import scipy.stats as stats
sns.jointplot(data3.Size,data3.Rating)
data3.annotate(stats.pearsonr)
plt.show()
```

```python
import scipy.stats as ss
#str1 will hold the value of Pearson r
str1 = 'Pearson r = {}'.format(ss.pearsonr(data3.Size,data3.Rating)[0].round(3))

#str2 will hold the p value
str2 = 'p value = {}'.format(ss.pearsonr(data3.Size,data3.Rating)[1].round(9))   
plt.figure(figsize = (10,10))
sns.jointplot(data3.Size, data3.Rating)

#Manually align the position of str1 and str2
plt.text(-1250,1.25, str1,horizontalalignment='center',verticalalignment='center',fontsize = 11 )
plt.text(-1250,1.0,str2,horizontalalignment='center',verticalalignment='center',fontsize = 11)
plt.show()
```

```python

str1 = 'Pearson r = {}'.format(ss.pearsonr(data3.Price,data3.Rating)[0].round(3))
str2 = 'p value = {}'.format(ss.pearsonr(data3.Price,data3.Rating)[1].round(9)) 
plt.figure(figsize = (10,10))
sns.jointplot(data3.Price, data3.Rating,kind='reg')
plt.text(-1250,1.25, str1,horizontalalignment='center',verticalalignment='center',fontsize = 11 )
plt.text(-1250,1.0,str2,horizontalalignment='center',verticalalignment='center',fontsize = 11)
plt.show()
```
